### CARGA DATAFRAME UNIDO 

```python
import xmlrpc.client
import pandas as pd

# Conexión
username = "juan.cano@donsson.com"
password = "1000285668"
url = "https://donsson.com"
db = "Donsson_produccion"

common = xmlrpc.client.ServerProxy(f"{url}/xmlrpc/2/common")
uid = common.authenticate(db, username, password, {})
models = xmlrpc.client.ServerProxy(f"{url}/xmlrpc/2/object")

# Campos a extraer
campos = [
    'producto_code',
    'almacen_id',
    'producto_costo_unitario',
    'clasificacion',
    'inventario_ideal',
    'producto_marca',
    'producto_code',
    'semana',
    'stock_actual',
    'product_name',
    'ano'
]

# Extraer en bloques
df_total = pd.DataFrame()
batch_size = 20000
offset = 0

while True:
    print(f"Cargando registros desde offset {offset}...")
    registros = models.execute_kw(
        db, uid, password,
        'cs.analisis.costo', 'search_read',
        [[]],  # Sin filtro de ciudad
        {'fields': campos, 'limit': batch_size, 'offset': offset}
    )
    
    if not registros:
        break

    df_parcial = pd.DataFrame(registros)
    df_total = pd.concat([df_total, df_parcial], ignore_index=True)
    offset += batch_size

    #demora en cargar aprox 30 seg
```

### EXPORTAR DATAFRAME 

```python
df_total.sample(10)
```

```python
# Asegurarte de que todos los valores sean strings simples
df_total['sucursal_nombre'] = df_total['almacen_id'].apply(lambda x: x[1] if isinstance(x, list) and len(x) > 1 else None)
df_total = df_total.drop(columns=['almacen_id', 'id'])
df_total =  df_total[['sucursal_nombre','producto_marca','product_name','producto_code','clasificacion','stock_actual','inventario_ideal','producto_costo_unitario','semana','ano']]

# Exportar

semana_unica = df_total["semana"].unique()

# Validación
if len(semana_unica) != 1:
    raise ValueError(f"Hay más de una semana en los datos: {semana_unica}")

n_semana = int(semana_unica[0])

# Guardar el archivo con el nombre de la semana
ruta_guardado = f'dataset/Entrada/df_completo_semana{n_semana}.csv'
df_total.to_csv(ruta_guardado, index=False)
print(f"Archivo guardado en: {ruta_guardado}")
```

```python
# Ahora sí agrupar
agrupado = df_total.groupby('sucursal_nombre')['stock_actual'].sum().reset_index()
agrupado = agrupado.sort_values(by='stock_actual', ascending=False)
print(agrupado)
```
