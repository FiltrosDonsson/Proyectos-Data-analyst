```python
duplicates = df_bald[df_bald.duplicated()].index

# 2. Drop those rows from the DataFrame
df_bald = df_bald.drop(duplicates)
```

```python
df_bald["name"] = "BALDWIN " + df_bald["Baldwin Number"].astype(str).str.upper().str.strip()
df_bald["name"] = df_bald["name"].str.replace("-", "", regex=False)


df_bald.head()
```

```python
# Filtrar Name que no est√©n en Odoo
df_bald_nuevos = df_bald[~df_bald["name"].isin(od_bald["name"])]
# Si df tambi√©n tiene Name
df_final = df_bald_nuevos.merge(od_bald, on="name", how="left")

df_final.info()
```

```python
# 1. Eliminar columnas que son solo NaN o tienen casi ning√∫n valor
# Mantendremos columnas que tengan al menos 1 valor no-NaN (si lo quieres m√°s estricto cambia 1 por 2, 3, etc.)
df_limpio = df_final.dropna(axis=1, how='all')

# 2. Definir y eliminar las columnas espec√≠ficas por su nombre exacto
cols_to_drop_specific = [
    "POP Code",
    "POP Code Change",
    "UPC - Each",
    "Qty/Case",
    "UPC - Case",
    "id","Name","Blue","List","Wt./Case"
]

# Filtrar solo las que existen
cols_exist = [col for col in cols_to_drop_specific if col in df_limpio.columns]

# Aplicar la eliminaci√≥n de las columnas espec√≠ficas
df_limpio = df_limpio.drop(columns=cols_exist)
```

```python
df_limpio.columns
```

```python
# Crear el nuevo DataFrame para la carga
df_cargar = df_limpio.copy()

# 1. Renombrar y manejar columnas existentes

# Renombrar 'name' a 'Name'
if 'name' in df_cargar.columns:
    df_cargar = df_cargar.rename(columns={'name': 'Name'})
else:
    # Si la columna 'name' no existe, puede ser 'id' o 'Baldwin Number' la que quieres renombrar
    print("Advertencia: La columna 'name' no se encontr√≥ para renombrar a 'Name'.")

# Renombrar '0.57' a 'Cost' (asumiendo que '0.57' es el nombre de columna correcto)
if 0.57 in df_cargar.columns:
    df_cargar = df_cargar.rename(columns={0.57: 'Cost'})
else:
    print("Advertencia: La columna 0.57 (para Cost) no se encontr√≥.")

# Renombrar 'Description' a 'Sales description'
if 'Description' in df_cargar.columns:
    df_cargar = df_cargar.rename(columns={'Description': 'description_sale'})
else:
    print("Advertencia: La columna 'Description' no se encontr√≥ para renombrar a 'Sales description'.")


# 2. Creaci√≥n de nuevas columnas con valores y c√°lculos

# C√°lculo del Sales Price (Costo / 0.8)
if 'Cost' in df_cargar.columns:
    # Usamos .loc para evitar SettingWithCopyWarning
    df_cargar.loc[:, 'Sales Price'] = df_cargar['Cost'] / 0.8
else:
    print("Error: No se puede calcular 'Sales Price' porque 'Cost' (antes '0.57') no existe.")

# Columna Product Type
df_cargar.loc[:, 'Product Type'] = 'product'

# Columnas vac√≠as
df_cargar.loc[:, 'Internal Reference'] = ''
df_cargar.loc[:, 'Barcode'] = ''
df_cargar.loc[:, 'Weight'] = ''

# Columna de Categor√≠a de Producto (categ_id)
# Usaremos el formato correcto para Odoo: [ID, Nombre]
df_cargar.loc[:, 'categ_id'] = 'Filters / Other brands'

# 3. Reordenar las columnas

# Identificar la columna que debe ser la primera (Baldwin Number o similar)
# Suponemos que la primera columna actual (√≠ndice 0) es la referencia principal
first_col = df_cargar.columns[0]

# Definir el nuevo orden: Primera, Name, luego las dem√°s
cols = [first_col, 'Name'] + [col for col in df_cargar.columns if col not in [first_col, 'Name']]
df_cargar = df_cargar[cols]
```

```python
df_cargar.head()
```

```python
# Suponiendo que df_cargar ya tiene las columnas calculadas y con valores fijos
# 1. Renombrar las columnas de df_cargar a los nombres EXACTOS de la plantilla Odoo

df_cargar = df_cargar.rename(columns={
    'Name': 'name',                        # De 'Name' a 'name' (min√∫scula)
    'Cost': 'standard_price',              # De 'Cost' (antes '0.57') a 'standard_price'
    'Sales Price': 'list_price',           # De 'Sales Price' a 'list_price'
    'Sales description': 'description',

})

# 2. Reafirmar valores de Odoo con los nombres correctos (si no los creaste as√≠)
df_cargar.loc[:, 'detailed_type'] = 'Storable Product' # 'storable' es el valor t√©cnico para Producto Almacenable
df_cargar.loc[:, 'sale_ok'] = True
df_cargar.loc[:, 'purchase_ok'] = True
# El campo de Cuenta de Gasto necesita el ID o la referencia externa, el .id es clave:
df_cargar.loc[:, 'property_account_expense_id/id'] = 'account.1_cost_of_goods_sold' # Asumo que es el nombre XMLID o referencia externa de la cuenta 500000
df_cargar.loc[:, 'show_availability'] = 'True' # Aunque es una columna booleana, poner 'True' asegura que se muestre.

# 3. Eliminar la columna 'id' que viene de la plantilla, pues es un ID de Odoo que no usaremos
if 'id' in df_cargar.columns:
    df_cargar = df_cargar.drop(columns=['id'])

# 4. Establecer la referencia interna (default_code)
# Asumiendo que la primera columna de df_limpio (Baldwin Number) es tu referencia interna
df_cargar.loc[:, 'default_code'] = df_cargar.iloc[:, 0]
df_cargar.loc[:, 'Quantity On Hand'] = 0 # O el nombre t√©cnico que te d√© Odoo
```

```python
df_cargar = df_cargar.drop(columns=["Baldwin Number","Marca","Internal Reference","Barcode","Weight","default_code"])
```
