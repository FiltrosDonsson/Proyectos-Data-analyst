```python
import pandas as pd 
import numpy as np
```

```python
ruta = "/home/donsson/proyectos/MODELO ABASTECIMIENTO/csvsalidas/clasificacion_mia2025.csv"
df_ns = pd.read_csv(ruta)
```

```python
df_ns = df_ns.drop(columns=["Unnamed: 0"])
df = df_ns[~(df_ns["store_name"]=="0")]
```

```python
# Definimos las condiciones para la meta, basadas en la clasificación
condiciones_meta = [
    (df['Clasificacion'] == 'AAA'),
    (df['Clasificacion'] == 'A'),
    (df['Clasificacion'] == 'B'),
    (df['Clasificacion'] == 'C')
]

# Definimos los valores de la meta que se asignarán a cada clasificación
metas = [0.999, 0.98, 0.95, 0.50]

df = df.copy()

# Asignamos la nueva columna 'Meta' usando np.select
df.loc[:, 'Meta'] = np.select(condiciones_meta, metas, default=np.nan)
```
